```python
import os
import pandas as pd
import numpy as np
from PIL import Image

# Load metadata
metadata_path = "dataset/HAM10000_metadata.csv"
metadata = pd.read_csv(metadata_path)

# Define image folder path
image_folder_path = "dataset/HAM10000_new_folder"

# Initialize lists to store images and labels
images = []
labels = []

# Loop through the metadata and load images
for index, row in metadata.iterrows():
    image_id = row['image_id']
    label = row['dx']
    
    # Construct image file path
    image_path = os.path.join(image_folder_path, f"{image_id}.jpg")
    
    # Load image and convert to NumPy array
    image = np.array(Image.open(image_path))
    
    # Append image and label to the lists
    images.append(image)
    labels.append(label)

# Convert lists to NumPy arrays
images = np.array(images)
labels = np.array(labels)

# Define the dataset structure as a tuple
dataset = (images, labels)

# Print the first two elements of the dataset for verification
print(dataset[0][:2])  # Print the first two images
print(dataset[1][:2])  # Print the corresponding labels
```
